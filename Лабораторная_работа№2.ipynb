```python
import pandas as pd
import plotly.express as px
import plotly.io as pio

# Загрузка данных из файла Google Sheets по ссылке (без заголовков)
url = "https://docs.google.com/spreadsheets/d/17lJ-lhEI5v77hBkr97gzhTkxd7WKQPU6/export?format=xlsx"
data = pd.read_excel(url, header=None)

# Задание названий столбцов
data.columns = ["Фабрика_1", "Фабрика_2", "Разрез"]

# Создание интерактивного графика с использованием Plotly Express
fig = px.line(data, x=data.index, y=["Фабрика_1", "Фабрика_2", "Разрез"], title='Почасовое потребление электроэнергии', render_mode='svg')

# Сортировка индексов по возрастанию
fig.update_xaxes(rangeslider_visible=True, tickangle=0, type='linear', categoryorder='array', categoryarray=sorted(data.index))

# Отображение графика
fig.show()

# Сохранение графика как изображение с использованием plotly.io
pio.write_image(fig, 'consumption_plot.png', scale=2)

# Сохранение данных в CSV файл
data.to_csv("consumption_data.csv", index=False)
df = pd.read_excel(url)
# вывод заголовка таблицы и первых пяти строк (сэмплов)
df.head()
print(df)
```
